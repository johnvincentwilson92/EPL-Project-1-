Is there a correlation between Goal Differential and Win %? 

```python
Project_1.columns
```

```python
Project_1 ["Date"] = pd.to_datetime(Project_1["Date"])
```

```python
Project_1
```

```python
Project_1 = Project_1[["Date", "HomeTeam", "AwayTeam", "FTHG", "FTAG", "FTR"]]
```

```python
Project_1.head()
```

```python
Project_1
```

```python
#Calling home win as calling from Numpy Home Goals greater than Away Goals, binary 1,0
Project_1["homewin"]= np.where(Project_1["FTHG"] > Project_1["FTAG"], 1,0)
```

```python
#Calling away win as calling from Numpy Away Goals greater than Home Goals, binary 1,0
Project_1["awaywin"] = np.where(Project_1["FTAG"] > Project_1 ["FTHG"], 1,0)
```

```python
# Calling a draw with the same logic above, using numpy Home Goals = Away Goals 
Project_1 ["draw"] = np.where(Project_1["FTHG"] == Project_1 ["FTAG"], .5, 0)
```

```python
Project_1 ["Count"] = 1 
```

```python
Project_1 
```

```python
HomeTeamData =Project_1.groupby(["HomeTeam"])[["FTHG", "FTAG", "homewin", "Count"]].sum().reset_index().rename(columns= {"HomeTeam": "Team", "FTHG": "FTHG_H", "FTAG": "FTAG_H", "Count": "Count_H"})

HomeTeamData
```

```python
AwayTeamData = Project_1.groupby(["AwayTeam"])[["FTHG", "FTAG", "awaywin", "Count"]].sum().reset_index().rename(columns= {"AwayTeam": "Team", "FTHG":"FTHG_A", "FTAG": "FTAG_A", "Count": "Count_A"})

AwayTeamData
```

Replace HomeTeamData 
Replicate for AwayTeam 

```python
Project_1 = pd.merge(HomeTeamData, AwayTeamData, on= ["Team"])
```

```python
Project_1.head()
```

```python
Project_1["Wins"] = Project_1 ["homewin"] + Project_1 ["awaywin"]
Project_1 ["Games"] = Project_1 ["Count_H"] + Project_1["Count_A"]
Project_1 ["GoalsFor"] = Project_1 ["FTHG_H"] + Project_1 ["FTHG_A"] 
Project_1 ["GoalsAgainst"] = Project_1 ["FTHG_A"] + Project_1 ["FTAG_H"] 
```

```python
Project_1.head()
```

```python
Project_1["Win%"] = Project_1["Wins"] /Project_1 ["Games"]
Project_1["GoalDiff"] = Project_1["GoalsFor"] - Project_1["GoalsAgainst"]
```

```python
Project_1.head()
```

```python
Project_1 ["Win%"].astype (float).head()
```

```python
Project_1.sort_values(by= ["Win%"], ascending= False).head()
```
